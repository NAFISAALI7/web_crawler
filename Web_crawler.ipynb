```python
import requests
import csv
import json

def scrape_youtube_channels(search_query, num_channels):
    api_key = 'AIzaSyB7mwH4SfU6tWlTsYiAC062uA1WNfyFMc4'
    base_url = 'https://www.googleapis.com/youtube/v3/search'

    params = {
        'key': api_key,
        'part': 'snippet',
        'q': search_query,
        'type': 'channel',
        'maxResults': num_channels
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    # Check the structure of the response
    print(data)

    channels = []
    if 'items' in data:
        for item in data['items']:
            channel_id = item['id']['channelId']
            channel_link = f'https://www.youtube.com/channel/{channel_id}'
            channels.append(channel_link)
    else:
        print('No items found in the response.')

    return channels

search_query = "openinapp.co"
num_channels = 10000
channels = scrape_youtube_channels(search_query, num_channels)

csv_file = 'youtube_channels.csv'
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Channel Links'])
    writer.writerows([[channel] for channel in channels])

json_file = 'youtube_channels.json'
data = {'channels': channels}
with open(json_file, 'w') as file:
    json.dump(data, file)
```
